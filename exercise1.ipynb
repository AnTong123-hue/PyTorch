```python
import torch

print("PyTorch Version:", torch.__version__)
print("CUDA Available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU Name:", torch.cuda.get_device_name(0))
else:
    print("Uh oh, CUDA isn't available. We might need to check your NVIDIA drivers.")
```

```python
# Set seed
torch.manual_seed(7)
# Create a random tensor
rand_tensor = torch.rand(1, 1, 1, 10)

# Remove single dimensions
squeezed_tensor = rand_tensor.squeeze()

# Print
print(f"Random Tensor: {rand_tensor}")
print(f"Squeezed Tensor: {squeezed_tensor}")
```
